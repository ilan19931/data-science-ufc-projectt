```python
import pandas as pd
import numpy as np 
import math

import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats
```

```python
df_main = pd.read_csv("my-data-set.csv")
df_copy = df_main.copy()
```

```python
df_copy.dtypes
```

```python
#loser wins is object. i will convert it to int.
df_copy.loc[df_copy['loser wins'] == "13-15"] = 0

df_copy['loser wins'] = df_copy['loser wins'].astype(str).astype(int)

df_copy.describe()
```

```python
# fix winner age and loser age negative values.

# set negative age to none. to calc currectly mean age.
df_copy.loc[df_copy['winner age'] <= 0, ['winner age']] = None
df_copy.loc[df_copy['loser age'] <= 0, ['loser age']] = None

winner_mean_age = round(df_copy['winner age'].mean())
loser_mean_age = round(df_copy['loser age'].mean())

#fill none ages to mean value
df_copy['winner age'] = df_copy['winner age'].fillna(winner_mean_age)
df_copy['loser age'] = df_copy['loser age'].fillna(loser_mean_age)

df_copy.describe()
```

```python
#fix loser loses data
#fix negative values by removing them (because there are small amout  of them)
print(df_copy.loc[df_copy['loser loses'] < 0,['loser loses']].count())

#removing negative values
df_copy.drop(df_copy[df_copy['loser loses'] < 0].index, inplace=True)
df_copy.describe()
```

```python
# fix bad fighters heights values, like 0.

# set 0 height values to none. to calc currectly mean height.
df_copy.loc[df_copy['winner height'] == 0, ['winner height']] = None
df_copy.loc[df_copy['loser height'] == 0, ['loser height']] = None

winner_mean_height = round(df_copy['winner height'].mean())
loser_mean_height = round(df_copy['loser height'].mean())

#fill none heights to mean value
df_copy['winner height'] = df_copy['winner height'].fillna(winner_mean_height)
df_copy['loser height'] = df_copy['loser height'].fillna(loser_mean_height)


df_copy.describe()
```

```python
# fix bad fighters reaches values, like 0.

# set 0 reaches values to none. to calc currectly mean reach.
df_copy.loc[df_copy['winner reach'] == 0, ['winner reach']] = None
df_copy.loc[df_copy['loser reach'] == 0, ['loser reach']] = None

df_copy.loc[df_copy['winner leg reach'] == 0, ['winner leg reach']] = None
df_copy.loc[df_copy['loser leg reach'] == 0, ['loser leg reach']] = None


winner_mean_reach = round(df_copy['winner reach'].mean())
loser_mean_reach = round(df_copy['loser reach'].mean())
winner_mean_leg_reach = round(df_copy['winner leg reach'].mean())
loser_mean_leg_reach = round(df_copy['loser leg reach'].mean())

#fill none reaches to mean value
df_copy['winner reach'] = df_copy['winner reach'].fillna(winner_mean_reach)
df_copy['loser reach'] = df_copy['loser reach'].fillna(loser_mean_reach)
df_copy['winner leg reach'] = df_copy['winner leg reach'].fillna(winner_mean_leg_reach)
df_copy['loser leg reach'] = df_copy['loser leg reach'].fillna(loser_mean_leg_reach)


df_copy.describe()
```

```python
#fix win round none values with mean value.

df_copy['win_round'] = df_copy['win_round'].fillna(round(df_copy['win_round'].mean()))
                                                   
df_copy.describe()
```

```python
#delete weight class NaN

df_copy.dropna(inplace=True)
```

```python
#clean method columns

df_copy['method'] = df_copy['method'].str.replace(r"\(.*?\)", "", regex=True)
df_copy['weight class'] = df_copy['weight class'].str.replace(r"\(.*?\)", "", regex=True)

df_copy.drop(df_copy[df_copy['win time'] == 0].index, inplace=True)
```

```python
#fix win time from string to number

def timeToNumber(time):
    timeArr = time.split(':')
    
    return int(timeArr[0]) * 60 + int(timeArr[1])
    
    
df_copy['win time'] = df_copy['win time'].apply(timeToNumber)
```

```python
#win round and win time fix

df_copy.loc[df_copy['win time'] > 300, ['win time']] = 300
df_copy.loc[df_copy['win_round'] > 5, ['win_round']] = 5


```

```python
#make categories

#fix method string

df_copy['method'] = df_copy['method'].str.replace(r"\[.*?\]", "", regex=True)
df_copy['method'] = df_copy['method'].apply(lambda x: x.lower())
df_copy['method'] = df_copy['method'].apply(lambda x: x.strip())
df_copy.loc[df_copy['method'] == "n/c", ['method']] = "no contest"
df_copy.loc[df_copy['method'] == "technical decision", ['method']] = "decision"
df_copy.loc[df_copy['method'] == "technical submission", ['method']] = "submission"
df_copy.loc[df_copy['method'] == "dq", ['method']] = "disqualification"
df_copy.loc[df_copy['method'] == "unanimous decision", ['method']] = "decision"

 

#fix weight class string
df_copy['weight class'] = df_copy['weight class'].apply(lambda x: x.lower())
df_copy['weight class'] = df_copy['weight class'].apply(lambda x: x.strip())



df_copy['method'] = pd.CategoricalIndex(df_copy['method'])
df_copy['weight class'] = pd.CategoricalIndex(df_copy['weight class'])

df_copy['method category'] = df_copy['method'].cat.codes
df_copy['weight class category'] = df_copy['weight class'].cat.codes

df_copy.head()
```

```python
#save to csv file

df_copy.to_csv("my-data-set-cleared.csv")
```
